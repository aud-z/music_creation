```python
import pandas as pd
import numpy as np
```

```python
tracks = pd.read_csv('tracks_features_qx.csv')
```

```python
artists = pd.read_table('v3_artist_ids.txt', sep = ';', names = ['artist', 'id'])
artists = artists['artist']
artists = list(artists)
'taylor swift' in artists
```

```python
print(tracks.count())
colNames = ["danceability", "energy", "key", "loudness", "mode",
             "speechiness", "acousticness", "instrumentalness",
             "liveness", "valence", "tempo", "duration_ms", "time_signature"]
df = tracks[colNames]
```

```python
print(df.count())
from sklearn import preprocessing
x = df.values 
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
```

```python
x_scaled.shape
```

```python
from itertools import combinations
from sklearn.cluster import KMeans
N = df.shape[0]
out = []

for j in range(3, 6):
    for comb in combinations(list(range(df.shape[1])), j):
        print(comb)
        cur_dat = x_scaled[:, comb]
        kmeans = KMeans(init="k-means++",
                n_init = 3,
                n_clusters= 15,
                random_state=15,
                max_iter = 200).fit(cur_dat)
        out.append(kmeans.labels_)
```

```python
out_array = np.array(out)
print(out_array.shape)
tracks.to_csv("tracks_qx.csv")
np.save('song_clusters_qx.npy', out_array)
```
